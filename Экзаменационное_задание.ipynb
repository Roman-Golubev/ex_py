## Описание данных


| Название колонки | Описание                                                                                                                                |
| ---------------- | --------------------------------------------------------------------------------------------------------------------------------------- |
| **absences**     | Количество пропущенных занятий                                                                                                          |
| **address**      | Тип адреса ученика ('U' — городской, 'R' — за городом)                                                                                  |
| **age**          | Возраст ученика (от 15 до 22 лет)                                                                                                       |
| **activities**   | Дополнительные внеучебные занятия (yes / no)                                                                                            |
| **Fedu**         | Образование отца (0 — нет, 1 — 4 класса, 2 — 5–9 классы, 3 — среднее специальное/11 классов, 4 — высшее)                                |
| **Fjob**         | Работа отца ('teacher' — учитель, 'health' — здравоохранение, 'services' — госслужба, 'at_home' — не работает, 'other' — другое)        |
| **famrel**       | Семейные отношения (от 1 — очень плохо до 5 — очень хорошо)                                                                             |
| **famsize**      | Размер семьи ('LE3' — ≤3 человека, 'GT3' — >3 человек)                                                                                  |
| **famsup**       | Семейная образовательная поддержка (yes / no)                                                                                           |
| **failures**     | Количество внеучебных неудач (n, если 1≤n≤3, иначе 0)                                                                                   |
| **free time**    | Свободное время после школы (от 1 — очень мало до 5 — очень много)                                                                      |
| **goout**        | Проведение времени с друзьями (от 1 — очень мало до 5 — очень много)                                                                    |
| **guardian**     | Опекун ('mother' — мать, 'father' — отец, 'other' — другое)                                                                             |
| **health**       | Текущее состояние здоровья (от 1 — очень плохо до 5 — очень хорошо)                                                                     |
| **higher**       | Хочет получить высшее образование (yes / no)                                                                                            |
| **internet**     | Наличие интернета дома (yes / no)                                                                                                       |
| **Medu**         | Образование матери (0 — нет, 1 — 4 класса, 2 — 5–9 классы, 3 — среднее специальное/11 классов, 4 — высшее)                              |
| **Mjob**         | Работа матери ('teacher' — учитель, 'health' — здравоохранение, 'services' — госслужба, 'at_home' — не работает, 'other' — другое)      |
| **nursery**      | Посещал детский сад (yes / no)                                                                                                          |
| **paid**         | Дополнительные платные занятия по математике (yes / no)                                                                                 |                                                                   |
| **Pstatus**      | Статус совместного проживания родителей ('T' — вместе, 'A' — раздельно)                                                                 |
| **reason**       | Причина выбора школы ('home' — близость к дому, 'reputation' — репутация школы, 'course' — образовательная программа, 'other' — другое) |
| **romantic**     | В романтических отношениях (yes / no)                                                                                                   |
| **school**       | Аббревиатура школы, в которой учится ученик                                                                                             |
| **schoolsup**    | Дополнительная образовательная поддержка (yes / no)                                                                                     |
| **score**        | Баллы по госэкзамену по математике                                                                                                      |
| **sex**          | Пол ученика ('F' — женский, 'M' — мужской)                                                                                              |
| **studytime**    | Время на учебу помимо школы в неделю (1 — <2 часов, 2 — 2–5 часов, 3 — 5–10 часов, 4 — >10 часов)                                       |
| **traveltime**   | Время в пути до школы (1 — <15 мин., 2 — 15–30 мин., 3 — 30–60 мин., 4 — >60 мин.)                                                      |
